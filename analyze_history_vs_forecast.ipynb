```python
import datetime, os
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
ipynb_path = os.path.dirname(os.path.realpath("__file__"))
```

# Load forecast data 

```python
data_file = os.path.join(ipynb_path, "output", "2021_04_09_10_17_Uhr.csv")

forecast_data1 = pd.read_csv(data_file, index_col="date")
forecast_data1.index = pd.to_datetime(forecast_data1.index)

data_file = os.path.join(ipynb_path, "output", "2021_04_09_20_50_Uhr.csv")

forecast_data2 = pd.read_csv(data_file, index_col="date")
forecast_data2.index = pd.to_datetime(forecast_data2.index)
```

# Load Historic data


```python
data_file = os.path.join(ipynb_path, "output", "2021_04_09_20_51_Uhr.csv")

historic_data = pd.read_csv(data_file, index_col="date")
historic_data.index = pd.to_datetime(historic_data.index)
```

# Comparison
## GHI

```python
fig, ax = plt.subplots()
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax.set_xlabel("Time") 
ax.set_ylabel("Global Irradiance [W/m2]")
ax.set_title("History vs. Forecast")

sns.lineplot(x=forecast_data1.index, y="RAD_WH", data=forecast_data1, label="Forecast 6 Uhr", ax=ax)
sns.lineplot(x=forecast_data2.index, y="RAD_WH", data=forecast_data2, label="Forecast 12 Uhr", ax=ax)
sns.lineplot(x=historic_data.index, y="RAD_WH", data=historic_data, label="History", ax=ax)
```

## Diffuse Radiation
